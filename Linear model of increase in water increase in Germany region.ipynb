```python
%%R
## regresión lineal simple entre Año Y Nivel del Mar
#
modlin<-lm(dados$Año~dados$NivelMar)
summary(modlin)
#
```

```python
%%R
## estimando a variancia del error
#
preditos<-modlin$fitted #valores predichos para el conjunto de dados directamente de R
residuos<-modlin$residuals #valores de los resíduos para el conjunto de datos directamente de R
mean(residuos)
sum(residuos^2)
#
```

```python
%%R
# matriz de variancia y covariancia de los betas
#
X<-cbind(1,dados$Año)
XTXinv<-solve(t(X)%*%X)
round(XTXinv,3)
summary(modlin)
#
```

```python
%%R
### análisis de diagnóstico
#
plot(dados$Año,residuos,xlab="Año")
qqnorm(residuos)
#
```

```python
%%R
plot(preditos,residuos) # assumindo que a ordem dos dados na tabela é a ordem de coleta, parece que temos pontos aberrantes e provavelmente influentes, pois temos menos resíduos negativos e muito mais positivos e resíduos negativos são valores bem mais distantes de zero.
#residuos
#dados
#
```

```python
## Intervalos de confianza 95%
#
%%R
coefs<-coefficients(modlin) #ou modlin[[1]]
des_pad<-coef(summary(modlin))[, "Std. Error"]
#
LI<-coefs-qt(0.975,16)*des_pad
LS<-coefs+qt(0.975,16)*des_pad
cbind(LI,LS)
```
